# Can we determine the best-performing foreign player in the 2020 Indian Premier League season?

```python
#IMPORT FUNCTIONS (NumPy, Panda, Math, MatPlotLib)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

######################################################################################################

#Basic Statistical Functions:

def standardDeviation(xList, xAvg):
    n = len(xList)
    numerator = 0
    for i in range(n):
        if math.isnan(xList[i]) == False:
            numerator = numerator + ((xList[i] - xAvg)**2)
    fraction = (1.0 * numerator) / n
    stdDev = np.sqrt(fraction)
    return stdDev
    
def square(x):
    return x * x

def divideTwoLists(xList, yList):
    if len(xList) == len(yList):
        zList = [None] * len(xList)
        for i in range(len(xList)):
            quotient = xList[i] / yList[i]
            zList[i] = quotient
    return zList

def findMean(lst):
    avg = 0
    n = len(lst)
    for value in lst:
        if math.isnan(value) == False:
            avg = avg + value
    avg = (1.0 * avg) / n
    return avg

def findZScore(lst, avg, sd):
    zSL = [None] * len(lst)
    for i in range(len(lst)):
        term = lst[i] - avg
        zScore = term / sd
        zSL[i] = zScore
        
    return zSL
    
def pearsonNum(xList, yList):
    if len(xList) == len(yList):
        n = len(xList)
        xyList = [None] * n
        for i in range(n):
            xyList[i] = xList[i] * yList[i]
        numerator = (n * np.sum(xyList)) - np.sum(xList) - np.sum(yList)
        return numerator
    else:
        return "There's Been An Error!"
    
def pearsonDenom(xList, yList):
    if len(xList) == len(yList):
        n = len(xList)
        xListS = sum(map(lambda x: square(x), xList))
        yListS = sum(map(lambda y: square(y), yList))
        
        xLeft = n * (xListS)
        yLeft = n * (yListS)
        
        xRight = square(np.sum(xList))
        yRight = square(np.sum(yList))
        
        denominator = (xLeft - xRight) * (yLeft - yRight)
        return denominator        
    else:
        return "There's Been An Error!"
    
def pearsonCoeff(xList, yList):
    return pearsonNum(xList, yList) / pearsonDenom(xList, yList)
    
```

# Adding Table of Foreign Players 

Input file is a CSV file containing statistics on every foreign player involved in 2020 IPL season. Due to the COVID-19 global pandemic, and other extraneous factors, many players pulled out, so I assigned a new table to contain the statistics of players that played in at least 5 games.

```python
csvURL = 'https://raw.githubusercontent.com/gprasad125/cricketProject/main/fullStatistics.csv'
foreignPlayers = pd.read_csv(csvURL)

activePlayers = foreignPlayers[foreignPlayers["Matches Played"] > 5]
activePlayers
```

Here are the statistics measured for each player, and what type of data type they are measured in.

Some important cricket-specific notes:
    - A player's Batting Average is calculated as (total # of runs scored / total # of innings the player got out)
    - A player's Strike Rate is calculated as (total # of runs scored / total # of balls faced) 
    - An "over" is a set of six balls, delivered by one bowler at a time.
    - A player's Bowling Average is calculated as (total # of wickets taken / total # of runs conceded)
    - A player's Economy Rate is calculated as (total # of runs conceded / total # of overs delivered)
    - A player's Bowling SR is calculated as (total # of wickets taken / total # of balls delivered)
    
Some of these statistics, such as balls delievered, or innings a player got out, were not included in the dataset, because other statistics utilize these statistics as part of their calculation.

Below we can see the different statistics measured, and the types of data values they carry. 

```python
activePlayers.dtypes
```

# Defining Batsmen, Bowlers, and All-Rounders

Cricketers perform one of three distinct roles in the game. They can be batsmen (those who score runs for the team), bowlers (those who take wickets and try and limit the other team's score), or all-rounders (those who can both bat and bowl serviceably). 

Below, we use the table's data to divide the players into the categories they perform:
- Batsman are defined as "Players who spent over 70% of their total matches batting, and less than 30% of their total matches bowling." We also will drop the bowling statistics for these players as we are only interested in their batting statistics.


- Bowlers are defined as "Players who spent over 70% of their total matches bowling, and less than 50% of their total matches batting." We are taking this 50% cutoff for the greater likelihood than any bowler will bat, as compared to any batter bowling. We will also drop the batting statistics for these players, as we are only interested in their batting statistics. 


- All-Rounders are defined as "Players who spent over 50% of their total matches bowling, and over 50% of their total matches batting." We will not drop any stats as we are interested in both types of statistics for these type of players.

Something that will help us define the above categories will be finding the percentage of matches played a particular player spent batting or bowling.

```python
matchesPlayed = activePlayers["Matches Played"].tolist()
battingInnings = activePlayers["Batting Innings"].tolist()
bowlingInnings = activePlayers["Bowling Innings"].tolist()

percentageBatting = divideTwoLists(battingInnings, matchesPlayed)
percentageBowling = divideTwoLists(bowlingInnings, matchesPlayed)

activePlayers.insert(5, "Percent of Matches Batting", percentageBatting)
activePlayers.insert(10, "Percent of Matches Bowling", percentageBowling)

activePlayers
```

Defining Batsmen

```python
batsmen = activePlayers.loc[(activePlayers["Percent of Matches Batting"] > 0.7) & (activePlayers["Percent of Matches Bowling"] < 0.3)].drop(["Bowling Innings","Percent of Matches Bowling", "Overs Delivered",  "Wickets", "Bowling Average", "Economy Rate", "Bowling SR"], axis = 1)
batsmen
```

Defining Bowlers

```python
bowlers = activePlayers.loc[(activePlayers["Percent of Matches Bowling"] > 0.7) & (activePlayers["Percent of Matches Batting"] < 0.5)].drop(["Batting Innings", "Percent of Matches Batting", "Runs Scored","Batting Average", "Strike Rate"], axis = 1)
bowlers
```

Defining All-Rounders

```python
allRounders = activePlayers.loc[(activePlayers["Percent of Matches Batting"] >= 0.5) & (activePlayers["Percent of Matches Bowling"] >= 0.5)]
allRounders
```

Checking that all players are accounted for

```python
#Getting the individual names for each category: Batsmen, Bowlers, All-Rounders
batNames = batsmen["Player Name"].tolist()
bowlNames = bowlers["Player Name"].tolist()
allRNames = allRounders["Player Name"].tolist()

#Getting all the names from the full table
names = activePlayers["Player Name"].tolist()

#Checking each category's list of names vs the total list of names by removing duplicates

count = len(names) #the number of active players 

for name in names:
    if name in batNames:
        count = count - 1
    elif name in bowlNames:
        count = count - 1
    elif name in allRNames:
        count = count - 1

#If all names from the full table are accounted and no player appears in two tables, this should yield True
count == 0
```

# Batsmen Calculations

Let's begin with the calculations for the batsmen. There are three categories we measure:
    - Runs Scored, a measure of impact: Did this player contribute a large amount to the team's overall season?
    - Batting Average, a measure of consistency: Did this player perform well in most matches?
    - Strike Rate, a measure of efficiency: How quickly did this player score his runs?
    
A good place to start would be to calculate the means and medians of each value.

```python
meanRuns = findMean(batsmen["Runs Scored"].tolist())
meanBatAvg = findMean(batsmen["Batting Average"].tolist())
meanSR = findMean(batsmen["Strike Rate"].tolist())

means = [meanRuns, meanBatAvg, meanSR]
means
```

We can use the mean to calculate the Standard Deviation and find which players performed well-above average in all three recorded statistics.

```python
sdRuns = standardDeviation(batsmen["Runs Scored"].tolist(), means[0])
sdBatAvg = standardDeviation(batsmen["Batting Average"].tolist(), means[1])
sdSR = standardDeviation(batsmen["Strike Rate"].tolist(), means[2])

stdDevs = [sdRuns, sdBatAvg, sdSR]
stdDevs
```

Now that we have the Standard Deviations, we can add a column to the table that tells roughly how many SDs above the mean someone's particular score is using a z-Score measure. 

```python
zScoreRuns = findZScore(batsmen["Runs Scored"].tolist(), means[0], stdDevs[0])
zScoreBA = findZScore(batsmen["Batting Average"].tolist(), means[1], stdDevs[1])
zScoreSR = findZScore(batsmen["Strike Rate"].tolist(), means[2], stdDevs[2])

batsmen.insert(9, "Runs Scored Z Scores", zScoreRuns)
batsmen.insert(10, "Batting Avg. Z Scores", zScoreBA)
batsmen.insert(11, "Strike Rates Z Scores", zScoreSR)

batsmen
```

Now that we know the zScores for each of our three categories, we can eliminate batsmen who have negative zScores in any of the categories, since they performed below average in the 2020 season.

```python
bestBats = batsmen.loc[(batsmen["Runs Scored Z Scores"] > 0) & (batsmen["Batting Avg. Z Scores"] > 0) & (batsmen["Strike Rates Z Scores"] > 0)]
bestBats
```

These 4 batsmen are the players who performed above-average in all measured categories. However, we want to find the best batsman, so let's see which players were had the highest zScores in the most categories.

```python
mostRuns = bestBats["Player Name"].tolist()[bestBats["Runs Scored Z Scores"].tolist().index(max(bestBats["Runs Scored Z Scores"].tolist()))]
topBatAvg = bestBats["Player Name"].tolist()[bestBats["Batting Avg. Z Scores"].tolist().index(max(bestBats["Batting Avg. Z Scores"].tolist()))]
topSR = bestBats["Player Name"].tolist()[bestBats["Strike Rates Z Scores"].tolist().index(max(bestBats["Strike Rates Z Scores"].tolist()))]
[mostRuns, topBatAvg, topSR]
```

We can see that AB de Villiers had the best SR and best Batting Avg. relative to the other top batsmen, so it makes sense to pick him as the best batsman. Furthermore, in the category he was not the best in: "Runs Scored", he had played less games than Quinton de Kock, the player with the most runs amongst the top batsmen, so it makes sense.

```python
abDeVilliers = bestBats.loc[(bestBats["Player Name"] == 'AB de Villiers')]
abDeVilliers
```

# Bowlers Calculations

Let's begin with the calculations for the bowlers. There are four categories we measure:
    
    - Wickets, a measure of impact: How much of an impact did this player overall have for his team?
    - Bowling Average, a measure of consistency: How well did this player perform each match he played in?
    - Economy Rate, Bowling SR, measures of efficiency: How quickly did this player perform vs opposition bats?
    
We can follow the same pattern we did for the batsmen: Find the mean, calculate standard deviation, calculate z-Scores, and find the players with positive z-Scores.

```python
meanWickets = findMean(bowlers["Wickets"].tolist())
meanBowlAvg = findMean(bowlers["Bowling Average"].tolist())
meanEconomy = findMean(bowlers["Economy Rate"].tolist())
meanBowlSR = findMean(bowlers["Bowling SR"].tolist())

meanBowl = [meanWickets, meanBowlAvg, meanEconomy, meanBowlSR]
meanBowl
```

```python
sdWickets = standardDeviation(bowlers["Wickets"].tolist(), meanBowl[0])
sdBowlAvg = standardDeviation(bowlers["Bowling Average"].tolist(), meanBowl[1])
sdEconomy = standardDeviation(bowlers["Economy Rate"].tolist(), meanBowl[2])
sdBowlSR = standardDeviation(bowlers["Bowling SR"].tolist(), meanBowl[3])

stdDevBowl = [sdWickets, sdBowlAvg, sdEconomy, sdBowlSR]
stdDevBowl
```

```python
zScoreWickets = findZScore(bowlers["Wickets"].tolist(), meanBowl[0], stdDevBowl[0])
zScoreBowlAvg = findZScore(bowlers["Bowling Average"].tolist(), meanBowl[1], stdDevBowl[1])
zScoreEconomy = findZScore(bowlers["Economy Rate"].tolist(), meanBowl[2], stdDevBowl[2])
zScoreBowlSR = findZScore(bowlers["Bowling SR"].tolist(), meanBowl[3], stdDevBowl[3])

bowlers.insert(11, "Wickets Z Scores", zScoreWickets)
bowlers.insert(12, "Bowling Avg. Z Scores", zScoreBowlAvg)
bowlers.insert(13, "Economy Rate Z Scores", zScoreEconomy)
bowlers.insert(14, "Bowling SR Z Scores", zScoreBowlSR)

bowlers
```

Note that we want to have negative z-Scores in the statistics of "Bowling Average", "Economy Rate", and "Bowling SR" because the lower these values are, the better a bowler is performing, so we want players with "lower than average" zScores. For example, a bowler with a bowling average of 20 (meaning he concedes 20 runs for 1 wicket) is far better than a bowler with a bowling average of 50. 

```python
bestBowl = bowlers.loc[(bowlers["Wickets Z Scores"] > 0) & (bowlers["Bowling Avg. Z Scores"] < 0) & (bowlers["Economy Rate Z Scores"] < 0) & (bowlers["Bowling SR Z Scores"] < 0)]
bestBowl
```

These 2 batsmen are the players who performed above-average in all measured categories. Therefore, let's find the player with the best zScores in each categories. 

```python
mostWickets = bestBowl["Player Name"].tolist()[bestBowl["Wickets Z Scores"].tolist().index(max(bestBowl["Wickets Z Scores"].tolist()))]
bestBowlAvg = bestBowl["Player Name"].tolist()[bestBowl["Bowling Avg. Z Scores"].tolist().index(min(bestBowl["Bowling Avg. Z Scores"].tolist()))]
bestEconRate = bestBowl["Player Name"].tolist()[bestBowl["Economy Rate Z Scores"].tolist().index(min(bestBowl["Economy Rate Z Scores"].tolist()))]
bestBowlSR = bestBowl["Player Name"].tolist()[bestBowl["Bowling SR Z Scores"].tolist().index(min(bestBowl["Bowling SR Z Scores"].tolist()))]

[mostWickets, bestBowlAvg, bestEconRate, bestBowlSR]
```

# All-Rounders Calculations

For all-rounders, we will repeat the steps we performed for batsmen and bowlers. 

```python
arMeanRuns = findMean(allRounders["Runs Scored"].tolist())
arMeanBatAvg = findMean(allRounders["Batting Average"].tolist())
arMeanSR = findMean(allRounders["Strike Rate"].tolist())
arMeanWickets = findMean(allRounders["Wickets"].tolist())
arMeanBowlAvg = findMean(allRounders["Bowling Average"].tolist())
arMeanEcon = findMean(allRounders["Economy Rate"].tolist())
arMeanBowlSR = findMean(allRounders["Bowling SR"].tolist())

arMeans = [arMeanRuns, arMeanBatAvg, arMeanSR, arMeanWickets, arMeanBowlAvg, arMeanEcon, arMeanBowlSR]

arSDRuns = standardDeviation(allRounders["Runs Scored"].tolist(), arMeans[0])
arSDBatAvg = standardDeviation(allRounders["Batting Average"].tolist(), arMeans[1])
arSDSR = standardDeviation(allRounders["Strike Rate"].tolist(), arMeans[2])
arSDWickets = standardDeviation(allRounders["Wickets"].tolist(), arMeans[3])
arSDBowlAvg = standardDeviation(allRounders["Bowling Average"].tolist(), arMeans[4])
arSDEcon = standardDeviation(allRounders["Economy Rate"].tolist(), arMeans[5])
arSDBowlSR = standardDeviation(allRounders["Bowling SR"].tolist(), arMeans[6])

arSD = [arSDRuns, arSDBatAvg, arSDSR, arSDWickets, arSDBowlAvg, arSDEcon, arSDBowlSR]

arzScoreRuns = findZScore(allRounders["Runs Scored"].tolist(), arMeans[0], arSD[0])
arzScoreBatAvg = findZScore(allRounders["Batting Average"].tolist(), arMeans[1], arSD[1])
arzScoreSR = findZScore(allRounders["Strike Rate"].tolist(), arMeans[2], arSD[2])
arzScoreWickets = findZScore(allRounders["Wickets"].tolist(), arMeans[3], arSD[3])
arzScoreBowlAvg = findZScore(allRounders["Bowling Average"].tolist(), arMeans[4], arSD[4])
arzScoreEcon = findZScore(allRounders["Economy Rate"].tolist(), arMeans[5], arSD[5])
arzScoreBowlSR = findZScore(allRounders["Bowling SR"].tolist(), arMeans[6], arSD[6])

allRounders.insert(16, "Runs Scored Z Scores", arzScoreRuns)
allRounders.insert(17, "Batting Avg. Z Scores", arzScoreBatAvg)
allRounders.insert(18, "Strike Rates Z Scores", arzScoreSR)
allRounders.insert(19, "Wickets Z Scores", arzScoreWickets)
allRounders.insert(20, "Bowling Avg. Z Scores", arzScoreBowlAvg)
allRounders.insert(21, "Economy Rate Z Scores", arzScoreEcon)
allRounders.insert(22, "Bowling SR Z Scores", arzScoreBowlSR)
 
allRounders
```

```python
bestAR = allRounders.loc[(allRounders["Runs Scored Z Scores"] > 0) 
                         & (allRounders["Batting Avg. Z Scores"] > 0) 
                         & (allRounders["Strike Rates Z Scores"] > 0) 
                         & (allRounders["Wickets Z Scores"] > 0) 
                         & (allRounders["Bowling Avg. Z Scores"] < 0) 
                         & (allRounders["Economy Rate Z Scores"] < 0) 
                         & (allRounders["Bowling SR Z Scores"] < 0)]
bestAR
```

As we can see above, no all-Rounder performed above average in every statistic. So, let's split the table into those who performed better as batsmen and those who performed better as bowlers

```python
bestARBat = allRounders.loc[(allRounders["Runs Scored Z Scores"] > 0) 
                         & (allRounders["Batting Avg. Z Scores"] > 0) 
                         & (allRounders["Strike Rates Z Scores"] > 0)]
bestARBat
```

```python
bestARBowl = allRounders.loc[(allRounders["Wickets Z Scores"] > 0) 
                         & (allRounders["Bowling Avg. Z Scores"] < 0) 
                         & (allRounders["Economy Rate Z Scores"] < 0) 
                         & (allRounders["Bowling SR Z Scores"] < 0)]
bestARBowl
```
